## II. Matplotlib Case Study

### .plot : 그래프 그리기에 사용되는 함수

- list 인자 1개만 들어갈 경우
`plt.plot([ ])` $\rightarrow$ x = index , y = list[ ]
- list 인자 2개개 들어갈 경우
`plt.plot(x,y)`
- y = f(x)

```python
plt.plot([2,4,2,4,2]) # 실제 plotting을 하는 함수 y = x + 1 # plot 꺾은선 그래프
# x = index , y = [1,2,3,4,5]
plt.show()  #plt를 확인하는 명령
```

### .figsize : Figure(도면)의 크기 조절

- `plt.figure(figsize=(x,y))` figsize는 tuple자료형 인자
- 실제로는 한픽셀의 72px$*$x, 72px$*$y 사이즈의 도면을 선언

```python
plt.figure(figsize=(6,6))  # plotting을 할 도면을 선언

plt.plot([0,1,2,3,4])
plt.show()
```

### 2차함수 그래프 with plot()

#### ※ list를 이용해서 1차함수 y = x를 그릴 경우

```python
plt.plot([0,1,2,3,4])
plt.show()
```

#### numpy.array를 이용해서 1차 함수 그리기

```python
x = np.array([1,2,3,4,5]) # 정의역
y = np.array([1,4,8,16,25]) # f(x)

plt.plot(x,y)
plt.show()
```

#### cf) range(a,b,c) vs np.arange(a,b,c)

```python
# np.arange(a,b,c) # a ~ b-c까지 c간격으로 -> [a,b)
# cf) range(a,b,c) # a ~ b-c까지 c간격으로 단 c가 무조건 int -> [a,b)

print(np.arange(1,10,0.5))
print(list(range(1,10,1)))
```

```python
x = np.arange(-10,10,0.01) 

plt.plot(x, x**2)
plt.show()
```

### 그래프 디테일 표현하기

#### x,y 축 label 이름 붙이기

- `plt.xlabel("x_label_name")` x축 label 이름
- `plt.ylabel("y_label_name")` y축 label 이름

```python
x = np.arange(-10,10,0.01) 

plt.xlabel("x value")  # x축 라벨
plt.ylabel("y value") # y축 라벨

plt.plot(x, x**2)
plt.show()
```

#### 보여질 x,y의 범위(축) 설정하기

- `plt.axis([x_min, x_max, y_min, y_max])`

```python
x = np.arange(-10,10,0.01)
plt.xlabel("x value")
plt.ylabel("y value")

plt.axis([-5,5,0,25]) # [ x_min, x_max, y_min, y_max] 보여질 축(axis) 설정

plt.plot(x, x**2)
plt.show()
```

#### x,y축의 눈금 설정하기

- `plt.xticks([ ])` # x축의 눈금 설정 list값
- `plt.yticks([ ])` # y축의 눈금 설정 list값


```python
x = np.arange(-10,10,0.01) 
plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0,25])

plt.xticks([i for i in range(-5,6,1)]) # x축의 눈금 설정 -5, -4, .. 6 '-1'까지 -> range(a,b,c)
plt.yticks([i for i in range(0,27,3)]) # y축의 눈금 설정 0, 3, 6,...27'-1'까지 -> range(a,b,c)

plt.plot(x, x**2)
plt.show()
```

#### 그래프의 제목과 범례(값 이름) 표시

- `plt.title("graph_name)`: 그래프 타이틀 설정
- `plt.plot(x,y,label="value_name")` : 그래프 값의 이름 설정
- `plt.legend()` : 범례 표시

```python
x = np.arange(-10,10,0.01) 

plt.xlabel("x value")
plt.ylabel("y value")
plt.axis([-5,5,0,25])
plt.xticks([i for i in range(-5,6,1)])
plt.yticks([i for i in range(0,27,3)])

plt.title(" y = x^2 graph")
plt.plot(x, x**2, label="trend")
plt.legend()

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프 (Plot)

- `plt.plot()`
- **시계열** 그래프에 사용됨 (시간의 지남에 따라 결과값)

```python
x = np.arange(20) # 0~19
y = np.random.randint(0,20,20) # randint(a,b,c) -> a~b까지 c번 int의 난수 생성

plt.axis([0,20,0,20])  # y축을 20까지 보이게 하기
plt.yticks([i for i in range(0,25,5)])  #y축을 5단위로 값

plt.plot(x,y)
plt.show()
```

#### cf) np.random.randit(a,b,c) vs random.randint(a,b)

```python
# np.random.randit(a,b,c=1) : [a,b) 범위 중 c번 난수 추출
# random.randint(a,b) : [a,b] 범위 중 난수 추출

import numpy as np
import random

np.random.randint(1,2), random.randint(1,2)
```

### 산점도 (Scatter Plot)

- `plt.scatter()`
- x,y가 완전히 별개의 변수일때 **상관관계 및 분포**

```python
plt.scatter(x,y)
plt.show()
```

### 박스 그림 (Box Plot)

- `plt.boxplot()`
- 단일 수치형 데이터에 대한 **축약된 정보** (Q1,Q2,Q3, min, max)  
(Q1~3 : 사분위수, Q2 = median)

```python
plt.boxplot(y)

plt.title("Box plot of y")
plt.show()
```

```python
plt.boxplot((x,y)) # 인자를 tuple 형식

plt.title("Box plot of x & y")
plt.show()
```

### 막대 그래프 (Bar Plot)

- `plt.bar(x,hight=)`
- 범주형 데이터의 값과 그 값의 크기를 직사각형으로 나타낸 그림
- **빈도(도수)**를 나타낼 때 주로 사용


```python
plt.bar(x,y)

plt.xticks(np.arange(0,20,1))
plt.show()
```

### cf. 히스토그램 (histogram)

- `plt.hist()`
- **도수분포**를 직사각형에 막대형태로 나타낸 그림
- "계급(bins)"이라는 분류로 나타냄 즉 범주형 데이터로 구성 후 그림을 그림
- 값끼리 붙어 있다.

```python
plt.hist(y, bins=np.arange(0,22,2))

plt.xticks(np.arange(0,22,2))
plt.show(), y
```

### 원형 그래프 (Pie Chart)

- `plt.pie()`
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에서 비해서 **비율** 확인에 용이

```python
z = np.arange(100,500,100)

plt.pie(z, labels=['one','two','three','four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

Matplotlib를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리

### Seaborn Import 하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)

- `sns.kdeplot()`
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린그림

```python
# Histogram
x = np.arange(0, 22, 2)  # [0,22) -> 0 ~ 20
y = np.random.randint(0,20,20) # [0,20) 20번의 난수 추출

plt.hist(y, bins=x)
plt.show()
```

```python
# Kdeplot

sns.kdeplot(y, shade=True) # shade는 boolean값으로 True는 면에 색상 False는 빈
plt.show()
```

### 카운트 그림(Count Plot)

- `sns.countplot()`
- 범주형 column의 빈도수를 시각화 $\rightarrow$ Groupby 후의 도수를 측정하는 것과 동일

```python
vote_df = pd.DataFrame({'name':['A','B','C'], 'vote':[True,True,False]})

vote_df
```

```python
# matplotlib barplot

vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False,True],height=vote_count['name'])
plt.show()
```

```python
# sns countplot

sns.countplot(x=vote_df['vote'])
plt.show()
```

### 캣그림 (Cat Plot)

- `sns.catplot(x=,y=,data=,kind='strip')`
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x="WHO Region",y="Confirmed",data=covid)
s.fig.set_size_inches(10,6) # sns스타일의 figure(면적) 조절
plt.show()
```

```python
s = sns.catplot(x="WHO Region",y="Confirmed",data=covid,kind='violin')
s.fig.set_size_inches(10,6) # sns스타일의 figure(면적) 조절
plt.show()
```

### 스트립그림 (Strip Plot)

- `sns.stripplot(x=,y=,data=)`
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프

```python
sns.stripplot(x='WHO Region',y='Recovered',data=covid)

plt.show()
```

```python
# cf) swarmplot 
# 뭉친 점들을 구분해서 나타내준다.
sns.swarmplot(x='WHO Region',y='Recovered',data=covid)

plt.show()
```

### 히트맵 (Heatmap)

- `sns.heatmap()`
- 데이터의 행렬을 색상으로 표현해주는 그래프
- 가장 많이 사용하는 경우, 상관계수행렬 

```python
covid.corr() # .corr() -> dataframe을 행렬로 만들어줌
```

```python
sns.heatmap(covid.corr())

plt.show()
```
